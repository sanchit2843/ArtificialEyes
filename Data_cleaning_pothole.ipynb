```
data=pd.read_csv("pothole1.csv")
```

```
times=data.Time.unique()
```

```
new_data = pd.DataFrame()
for t in times:
  temp=data.loc[data['Time'] == t].shape[0]
  if(temp<20):
    continue
  temp_data=data.loc[data['Time'] == t].head(20)
  temp_data=temp_data.iloc[:,1:].to_numpy()
  temp_data=pd.DataFrame(np.reshape(temp_data,[1,120]))
  new_data=new_data.append(temp_data,ignore_index=True)
```
